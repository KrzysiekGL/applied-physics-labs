```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
func1 = pd.read_csv("func1.csv")
func2 = pd.read_csv("func2.csv")
func3 = pd.read_csv("func3.csv")
```

```python
fig, ax = plt.subplots()

ax.plot(func1.t, func1.theta, ".", label="Simple pendulum")
ax.plot(func2.t, func2.theta, ".", label="Dissipation")
ax.plot(func3.t, func3.theta, ".", label="Dissipation + Driving force")

legend = ax.legend(loc='upper right', fontsize='x-small')

ax.set_title("Pendulum - second order DE solved with RK method")
ax.set_xlabel("time [s]")
ax.set_ylabel("amplitude [rad]")

plt.show()
```
