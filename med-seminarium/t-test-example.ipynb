```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Próbka
n = 40
dof = n-1
xdash = 75
s = 15

# Poziom istotności
alpha = 0.01
ntails = 2

# Hipoteza zerowa
mu = 80
```

```python
ttest = (xdash - mu) * np.sqrt(n) / s
```

```python
tcrit = abs(st.t.ppf(alpha/ntails, dof)) # t-score table
```

```python
plt.figure()
xs = np.linspace(-10,10,1000)

# t-rozkład
plt.plot(xs, st.t.pdf(xs, dof), 'k')

# t krytyczne
plt.vlines([-tcrit, tcrit], 0.0, st.t.pdf(tcrit, dof), colors='r')

# t - wynik testu
y = st.t.pdf(ttest, dof)
if(y < 0.05): y = 0.05
plt.vlines(ttest, 0.0, y, colors='b')

# info
print("zmieniamy s z 5 na ", s)
print("i zmniejszamy alpha do", alpha)
print("n: ", n)
print("x: ", xdash)
print("s: ", s)
print("H0: mu = ", mu)
print("alpha:", alpha)
print("t test: ", ttest)
print("+/- t crit:", tcrit)
```
